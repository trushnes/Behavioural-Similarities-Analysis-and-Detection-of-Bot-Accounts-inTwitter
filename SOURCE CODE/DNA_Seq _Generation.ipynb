```python
col_list = ["user_id", "in_reply_to_user_id", "retweeted_status_id"]
df = pd.read_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/legit_user_b3_type.csv", usecols = col_list)
print(df.head())
```

```python
print(df.tail(10))
```

```python
for c in df.columns:
    df[c] = pd.to_numeric(df[c], errors='coerce')
```

```python
df.to_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/legit_user_b3_type.csv")
df = df.dropna()
```

```python
nrows = df['user_id'].nunique()
print(nrows)
```

```python
user_seq_list = []
seq_arr = []
user_id_arr = df['user_id'].unique()
user_group = df.groupby(['user_id'])
```

```python
print(user_id_arr[2])
print(len(user_id_arr))
```

```python
for i in range(0,nrows):
    s = ""
    curr_user_grp = user_group.get_group(user_id_arr[i])
    for row in curr_user_grp.itertuples():
        print(row.user_id)
        if row.in_reply_to_user_id != 0.0:
            s = s + "C"
        elif row.retweeted_status_id != 0.0:
            s = s + "T"
        else:
            s = s + "A"
    print(s)
    print("--------------------------")
    user_seq_list.append((user_id_arr[i], s))
    seq_arr.append(s)
```

```python
longest_seq = max(seq_arr, key = len)
print(len(longest_seq))
shortest_seq = min(seq_arr, key = len)
print(len(shortest_seq))
```

```python
seq_df = pd.DataFrame(user_seq_list, columns =['user_id', 'sequence'])
seq_df.to_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/legit_user_b3_type.csv")
```

**Extracting Digital DNA Sequence using B3 Content**

```python
import pandas as pd
data2 = pd.read_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/tweets.csv", usecols = [3,18,19,20])
data2.to_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/genuine_user_content_data.csv", header=['user_id', 'num_hashtags', 'num_urls', 'num_mentions'])

col_list2 = ["user_id", "num_hashtags", "num_urls", "num_mentions"]
df2 = pd.read_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/genuine_user_content_data.csv", usecols = col_list2)

print(df2.head())
```

```python
for c in df2.columns:
    df2[c] = pd.to_numeric(df2[c], errors='coerce')

#df2 = df2.dropna()
df2['num_hashtags'].fillna(0, inplace=True)
df2['num_urls'].fillna(0, inplace=True)
df2['num_mentions'].fillna(0, inplace=True)

df2.to_csv("genuine_user_content_data.csv")
df2 = df2.dropna()

nrows2 = df2['user_id'].nunique()
print(nrows2)
```

```python
user_content_seq_list = []
content_seq_arr = []
user_id_content_arr = df2['user_id'].unique()
user_content_group = df2.groupby(['user_id'])

print(user_id_content_arr[2])
print(len(user_id_content_arr))
```

```python
for i in range(0,nrows2):
    s = ""
    curr_user_grp = user_content_group.get_group(user_id_content_arr[i])
    for row in curr_user_grp.itertuples():
        print(row.user_id)
        if row.num_hashtags == 0.0 and row.num_urls == 0.0 and row.num_mentions == 0.0:
            s = s + "N"
        elif row.num_hashtags != 0.0 and row.num_urls == 0.0 and row.num_mentions == 0.0:
            s = s + "E"
        elif row.num_hashtags == 0.0 and row.num_urls != 0.0 and row.num_mentions == 0.0:
            s = s + "E"
        elif row.num_hashtags == 0.0 and row.num_urls == 0.0 and row.num_mentions != 0.0:
            s = s + "E"
        else:
            s = s + "X"
        
    print(s)
    print("--------------------------")
    user_content_seq_list.append((user_id_content_arr[i], s))
    content_seq_arr.append(s)
```

```python
longest_seq2 = max(content_seq_arr, key = len)
print(len(longest_seq2))
shortest_seq2 = min(content_seq_arr, key = len)
print(len(shortest_seq2))
```

```python
content_seq_df = pd.DataFrame(user_content_seq_list, columns =['user_id', 'sequence'])
content_seq_df.to_csv("/content/drive/MyDrive/Projects/datasets_full.csv/genuine_accounts.csv/genuine_user_content_b3_seq.csv")
```

### **Obtaining Longest Common Subsequence**

```python
def LCSSubStr(X, Y, m, n):
	LCSuffix = [[0 for i in range(n + 1)] 
				for j in range(m + 1)] 

	length = 0
	row, col = 0, 0

	for i in range(m + 1): 
		for j in range(n + 1): 
			if i == 0 or j == 0: 
				LCSuffix[i][j] = 0
			elif X[i - 1] == Y[j - 1]: 
				LCSuffix[i][j] = LCSuffix[i - 1][j - 1] + 1
				if length < LCSuffix[i][j]: 
					length = LCSuffix[i][j] 
					row = i 
					col = j 
			else: 
				LCSuffix[i][j] = 0

	if length == 0: 
		return ""
 
	resultStr = ['0'] * length 

	while LCSuffix[row][col] != 0: 
		length -= 1
		resultStr[length] = X[row - 1] 

		row -= 1
		col -= 1

	return (''.join(resultStr))
```

```python
def findstem(arr):
    n = len(arr)
    j = 1;
    X = arr[0]
    Y = arr[j]
    k = 2
    lcs_data = []
    
    while j < n:
        Y = arr[j]
        xlen = len(X)
        ylen = len(Y)
        xySubStr = LCSSubStr(X,Y,xlen,ylen)
        lcs_data.append((k, len(xySubStr)))
        print("\t\tProgress message: k = %d length of lcs = %d" %(k, len(xySubStr)))
        k = k+1
        j = j+1
        X = xySubStr
    
    return lcs_data
```

```python
def lcsProcessHandler(source_file_path, save_lcs_data_file_name):
    col_list = ["sequence"]
    seq_df = pd.read_csv(source_file_path, usecols = col_list)
    seq_list = seq_df['sequence'].tolist()
    seq_list.sort(key = len, reverse=True)
    print("================ Findins LCS data for "+source_file_path+" ================")
    lcs_data_list = findstem(seq_list)
    lcs_data_list_df = pd.DataFrame(lcs_data_list, columns =['num_of_strings', 'lcs_length'])
    lcs_data_list_df.to_csv(save_lcs_data_file_name)
    print("LCS process complete for "+source_file_path)
```
